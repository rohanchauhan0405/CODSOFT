Loading csv file to Data Frame

```python
#Loading  data into data frame
df=pd.read_csv('CreditCard.csv')
```

```python
# checking the Number of Missing values
# print(df.isnull().sum()) re turns the number of missing values

print(df.isnull().sum())
```

```python
# Dropping Missing Values
df=df.dropna()
```

```python
#after Dropping Missing values  checking again

print(df.isnull().sum())
```

```python
# use to  find the Duplicates from the data sets

df[df.duplicated()]
```

```python
# Printing the Number of Duplicate rows

duplicate_rows_df=df[df.duplicated()]
print("Number of Duplicate rows : ", duplicate_rows_df)
```

```python
# Dropping  the Duplicates

df=df.drop_duplicates()
df.head()
```

```python
# changing the Data Type  of  Class

df['Class']=df['Class'].astype(int)
```

```python
#  fot data analysis we seperating the  data
legit=df[df.Class ==0]
fraud=df[df.Class ==1]
```

Under Smapling

Building sample data set from  original  dataset  Containing similar to normal transactions  and fraudulent transactions

```python
legit_sample=legit.sample(n=73)
```

Concatinating Two Data  frames

```python
new_df=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
# Counting the Values of normal Transactions and Fraudlent Transactions from new data frame

new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
x=new_df.drop(columns='Class',axis=1)
y=new_df['Class']
```

```python
x
```

```python
y
```

Spliting data into training data and testing data

```python
# Here we Creating the 4 arrays x_train,x_test,y_train,y_test
# Here x represents the features and y represents the Labels
# all features are stored in x_train and x_test and Labels are stored in y_train and y_test

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2,stratify=y)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model Training

Logistic Regression

```python
model=LogisticRegression()
```

```python
#Training the logistic Regression model with Training Data

model.fit(x_train,y_train)
```

 model  Evaluation.

 Accuracy Score

```python
# Accuracy of Training data

x_train_prediction=model.predict(x_train)
accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print('The Accuracy is : ' ,accuracy)
```

```python
# accuracy of Test data

x_test_prediction=model.predict(x_test)
test_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy of Test Data : ',test_accuracy)
```
