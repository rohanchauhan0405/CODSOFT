Handling Outliers

```python
R1=df.quantile(0.25)
R2=df.quantile(0.75)
IQR=R2-R1
print(IQR)
```

```python
# here we used formula for handling outliers

df=df[~((df<(R1-1.5*IQR))|(df>(R2+1.5*IQR))).any(axis=1)]
df.shape
```

```python
sns.boxplot(x='Newspaper', data=df)
```

```python
#plotting a heatmap of Data frame

plt.figure(figsize=(10,5))
c=df.corr()
sns.heatmap(c,cmap="YlGnBu",annot=True)
c
```

```python
x=df.drop(columns='Sales')
y=df['Sales']
```

Model Training

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.2,test_size=0.3,random_state=2)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_test.head()
```

```python
y_train.head()
```

```python
# imporitng a LinearRegression

from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()

```

```python
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import  mean_absolute_error,mean_squared_error,r2_score
```

```python
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
print(model.score(x_train,y_train))
print(model.score(x_test,y_test))
```
